```python
# generates a sequence of lgth numbers based on n0 and n1

def generate(n0, n1, lgth):
    print(n0, n1, end=' ')
    for i in range(lgth - 2):
        n = n0 + n1
        print(str(n)+ ' ', end='')
        n0 = n1
        n1 = n
```

```python
# fibonacci

generate(0, 1, 16)
```

```python
# lucas

generate(2, 1, 16)
```

```python
# recursive fibonacci implementation
# (short and elegant, but catastrophically inefficient)

def fibonacci_r(term):
    print(f'calculating term {term}')
    if term == 0:
        return 0
    elif term == 1:
        return 1
    else:
        return fibonacci_r(term - 2) + fibonacci_r(term - 1)

fibonacci_term = 6
print(f'fibonacci_r({fibonacci_term}) = {fibonacci_r(fibonacci_term)}')
```
