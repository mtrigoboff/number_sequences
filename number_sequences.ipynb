```python
# generates a sequence of lgth numbers based on n0 and n1

def generate(n0, n1, lgth):
    print(n0, n1, end=' ')
    for i in range(lgth - 2):
        n = n0 + n1
        print(str(n)+ ' ', end='')
        n0 = n1
        n1 = n
```

```python
# fibonacci

generate(0, 1, 16)
```

```python
# lucas

generate(2, 1, 16)
```

```python
# recursive fibonacci implementation
# (short and elegant code, but catastrophically inefficient)

indent = 3          # number of spaces to indent each level

def fibonacci_r1(term, level=0):
    print(f'{' ' * indent * level}calculating term {term}')
    if term == 0:
        return 0
    elif term == 1:
        return 1
    else:
        return fibonacci_r1(term - 2, level + 1) + fibonacci_r1(term - 1, level + 1)

which_term = 6
fibonacci_term = fibonacci_r1(which_term)
print()
print(f'fibonacci_r1({which_term}) = {fibonacci_term}')
```

```python
# recursive fibonacci implementation
# (more efficient recursive code)

indent = 3          # number of spaces to indent each level

def fibonacci_r2(term_no, level):
    spacer = ' ' * indent * level
    print(f'{spacer}term({term_no}) called')
    if term_no == 0:
        terms = (0, 1)
    else:
        terms = fibonacci_r2(term_no - 1, level + 1)
        terms = (terms[1], terms[0] + terms[1])
    print(f'{spacer}term({term_no}) returning {terms}')
    return terms

def fibonacci_2(term_no):
    return fibonacci_r2(term_no, 0)[0]

term = 6
fib_term = fibonacci_2(6)
print()
print(f'fibonacci term({term}) = {fib_term}')
```
